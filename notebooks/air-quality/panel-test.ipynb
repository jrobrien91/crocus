```python
import panel as pn
import matplotlib.pyplot as plt
```

```python
pn.extension()

markdown = pn.pane.Markdown('Hello World')
markdown.servable()
```

```python
pn.extension()

markdown = pn.pane.Markdown("Hello World!")
card = pn.Card(markdown)
card.servable()
```

```python
pn.extension()

package = pn.widgets.Select(
    name="Package", options=["Panel", "Streamlit"]
)
column = pn.Column(package)

# set package value
package.value = "Streamlit"

# get package value
column.append(f"Selected package: {package.value}")

column.servable()
```

```python
pn.extension()

def callback(clicks):
    return pn.widgets.StaticText(name="Clicks", value=str(clicks))

button = pn.widgets.Button(name="Click me!")
counter = pn.bind(callback, clicks=button.param.clicks)

pn.Column(button, counter).servable()
```

```python
pn.extension()

template = pn.template.FastListTemplate(
    sidebar=["# Sidebar Column"],
    main=["# Main Column"],
    header=["# Header Column"],
)
template.servable()
```

# Making the Display

```python
slider = pn.widgets.IntSlider(name="Year",
                              value=1950,
                              start=1950,
                              end=2022,
                              step=1
                             )
name_select = pn.widgets.Select(name="Name")
plot_preview = pn.pane.Matplotlib(plt.figure(),
                                  sizing_mode='scale_both'
                                 )
sidebar = pn.Column(slider, name_select)
main = pn.Row(plot_preview)

template = pn.template.FastListTemplate(sidebar=sidebar,
                                        main=main
                                       )
```
