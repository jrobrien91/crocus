# Quality Control for the Vaisala Weather Transmitter (WXT-536)

[CROCUS](https://crocus-urban.org/) utilizes the Vasiala WXT-536 within all Chicago Micronet nodes. <br>
Data are collected at the instrument via [Waggle Plugin Scripts](https://github.com/jrobrien91/waggle-wxt536), which incorporate the sensors into the Waggle node. <br>
Data collected at each site are accumulated into daily files via [ingest scripts](https://github.com/CROCUS-Urban/ingests/blob/main/scripts/wxt-ingest.py) 
and hosted on the [CROCUS Thredds Server](https://thredds.crocus-urban.org/thredds/catalog/neiu-wxt-a1/catalog.html)
